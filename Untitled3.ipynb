```python
top5 = []

for i in range(5):
    max = 0
    max_index = 0
    for j in range(len(df)):
        # if int(df.iloc[j]['허가병상수']) in top5:
        #     continue
        if j in top5:
            continue
        if max < int(df.iloc[j]['허가병상수']):
            max = int(df.iloc[j]['허가병상수'])
            max_index = j
    top5.append(max_index)
print(top5)
```
